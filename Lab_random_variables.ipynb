# Lab: random variable distribution

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

##    Get the numerical variables from our dataset.


```python
data = pd.read_csv("files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv")
```

```python
num = data.select_dtypes(include = np.number)
```

```python
num.head()
```

##    Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.


```python
def plotter(df):
    for column in df: 
        sns.displot(df[column], kde = True)
plotter(num)
```

Customer lifetime value, number of open complaints, and number of policies are pretty clearly exponential. 

The only straightforward normal distribution seems to be 'months since policy inception', but there may well be bi- or trimodal normal distributions, e.g. in monthly premium auto and total claim amount. 

In 'income', the large number of zeroes muddles the picture a bit - there might be a bimodal normal distribution there as well.

##    Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.


```python
#I'm doing the sloppy log transformation, where I don't watch out for values of zero (that generate -inf scores). 
# But this is not a problem because the plots will ignore the -inf values anyway, so the dirty method leads to clean output in the end. 

numlog = num.apply(np.log)
```

```python
plotter(numlog)
```
