And the Policy column can be simplified by taking only L1, L2, and L3:

```python
newpolicy = []

for column in categorical_df['Policy']:
    if 'L1' in column: 
        newpolicy.append('L1')
    elif 'L2' in column: 
        newpolicy.append('L2')
    elif 'L3' in column:
        newpolicy.append('L3')

```
